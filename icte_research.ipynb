```python
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})
angles = np.linspace(0, 2 * np.pi, len(systems), endpoint=False).tolist()

values = kernel_extraction_times + kernel_extraction_times[:1]
angles += angles[:1]

ax.fill(angles, values, color='cyan', alpha=0.25)
ax.plot(angles, values, marker='o', color='navy')  # Line with markers

ax.set_xticks(angles[:-1])
ax.set_xticklabels(systems)

for label, angle, radius in zip(kernel_extraction_times, angles, values):
    ax.text(angle, radius + 0.5, f"{label} sec", horizontalalignment='center', size=12, color='black')

ax.set_title('Kernel Extraction Time (seconds) - Polar Plot')
plt.show()
```
